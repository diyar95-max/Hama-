```python
# %% [markdown]
# # Health and Wellness Data Preparation and Cleaning
#
# This notebook prepares and cleans the simulated health and wellness dataset for analysis.

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# %%
# Load the dataset
df = pd.read_csv('simulated_health_wellness_data.csv')

# Display basic information
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())

# %%
# Basic statistics
print("\nDescriptive statistics:")
display(df.describe())

# %%
# Check for missing values
print("\nMissing values per column:")
display(df.isnull().sum())

# %% [markdown]
# ## Data Cleaning

# %%
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())

# %%
# Check for impossible values
print("\nChecking for impossible values:")

# Exercise time should be between 0 and 1440 (minutes in a day)
print("Exercise time min:", df['Exercise_Time_Min'].min())
print("Exercise time max:", df['Exercise_Time_Min'].max())

# Healthy meals should be between 0 and 10 (assuming no one eats more than 10 healthy meals per day)
print("\nHealthy meals min:", df['Healthy_Meals_Per_Day'].min())
print("Healthy meals max:", df['Healthy_Meals_Per_Day'].max())

# Sleep hours should be between 1 and 16 (reasonable human sleep range)
print("\nSleep hours min:", df['Sleep_Hours_Per_Night'].min())
print("Sleep hours max:", df['Sleep_Hours_Per_Night'].max())

# Stress level should be between 1 and 10 (assuming 1-10 scale)
print("\nStress level min:", df['Stress_Level'].min())
print("Stress level max:", df['Stress_Level'].max())

# BMI should be between 10 and 50 (reasonable human BMI range)
print("\nBMI min:", df['BMI'].min())
print("BMI max:", df['BMI'].max())

# %%
# Handle any impossible values if found (none in this dataset)
# Example code for handling if we found any:
# df['Sleep_Hours_Per_Night'] = df['Sleep_Hours_Per_Night'].clip(1, 16)

# %%
# Check for outliers using boxplots
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
sns.boxplot(y=df['Exercise_Time_Min'])
plt.title('Exercise Time (min)')

plt.subplot(2, 3, 2)
sns.boxplot(y=df['Healthy_Meals_Per_Day'])
plt.title('Healthy Meals Per Day')

plt.subplot(2, 3, 3)
sns.boxplot(y=df['Sleep_Hours_Per_Night'])
plt.title('Sleep Hours Per Night')

plt.subplot(2, 3, 4)
sns.boxplot(y=df['Stress_Level'])
plt.title('Stress Level')

plt.subplot(2, 3, 5)
sns.boxplot(y=df['BMI'])
plt.title('BMI')

plt.tight_layout()
plt.show()

# %%
# Handle outliers - we'll cap extreme values at reasonable thresholds
# For this dataset, we'll leave them as they appear to be valid extreme cases

# %% [markdown]
# ## Feature Engineering

# %%
# Create some potentially useful derived features

# 1. Create a 'Total_Wellness_Score' combining multiple factors
# (This is a simplified example - in reality you'd want to normalize these first)
df['Total_Wellness_Score'] = (
    df['Exercise_Time_Min'] / 60 +  # Convert to hours
    df['Healthy_Meals_Per_Day'] +
    df['Sleep_Hours_Per_Night'] -
    df['Stress_Level']
)

# 2. Create BMI categories
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['BMI'].apply(bmi_category)

# 3. Create sleep adequacy flag
df['Adequate_Sleep'] = df['Sleep_Hours_Per_Night'].apply(lambda x: 1 if x >= 7 else 0)

# %%
# Display the new features
print("\nDataset with new features:")
display(df.head())

# %% [markdown]
# ## Data Quality Check

# %%
# Check correlation between variables
plt.figure(figsize=(10, 8))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# %%
# Check distribution of categorical variables
print("\nBMI Category distribution:")
display(df['BMI_Category'].value_counts())

# %%
# Check for any illogical relationships
# For example, high exercise but very high BMI might be suspicious
high_exercise_high_bmi = df[(df['Exercise_Time_Min'] > df['Exercise_Time_Min'].quantile(0.75)) &
                           (df['BMI'] > df['BMI'].quantile(0.75))]

print("\nHigh exercise high BMI cases:")
display(high_exercise_high_bmi)

# %% [markdown]
# ## Save Cleaned Data

# %%
# Save the cleaned dataset
df.to_csv('cleaned_health_wellness_data.csv', index=False)

print("\nCleaned dataset saved successfully!")
```

```python
# %% [markdown]
# # Descriptive Data Analysis - Health and Wellness Data

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set up visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# Load the cleaned dataset
df = pd.read_csv('cleaned_health_wellness_data.csv')

# %% [markdown]
# ## 1. Basic Dataset Overview

# %%
# Display basic information
print("="*50)
print("Dataset Overview")
print("="*50)
print(f"Number of observations: {df.shape[0]}")
print(f"Number of variables: {df.shape[1]}")
print("\nFirst 5 rows:")
display(df.head())

print("\nData types:")
display(df.dtypes)

print("\nMissing values per column:")
display(df.isnull().sum())

# %% [markdown]
# ## 2. Univariate Analysis

# %%
# Select numerical columns
numerical_cols = ['Exercise_Time_Min', 'Healthy_Meals_Per_Day',
                 'Sleep_Hours_Per_Night', 'Stress_Level', 'BMI',
                 'Total_Wellness_Score']

# Select categorical columns
categorical_cols = ['BMI_Category', 'Adequate_Sleep']

# %%
print("="*50)
print("Descriptive Statistics for Numerical Variables")
print("="*50)
display(df[numerical_cols].describe().T)

# %%
# Plot distributions for numerical variables
print("\nDistribution Plots:")
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# %%
# Plot boxplots for numerical variables to check for outliers
print("\nBoxplots for Numerical Variables:")
plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()

# %%
# Normality tests for numerical variables
print("\nNormality Tests (Shapiro-Wilk):")
normality_results = []
for col in numerical_cols:
    stat, p = stats.shapiro(df[col])
    normality_results.append({
        'Variable': col,
        'W-statistic': stat,
        'p-value': p,
        'Normal': p > 0.05
    })

normality_df = pd.DataFrame(normality_results)
display(normality_df)

# %%
# Analysis for categorical variables
print("="*50)
print("Descriptive Statistics for Categorical Variables")
print("="*50)

for col in categorical_cols:
    print(f"\nFrequency table for {col}:")
    display(pd.DataFrame(df[col].value_counts()).T)

    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()

# %% [markdown]
# ## 3. Bivariate Analysis

# %%
# Correlation analysis
print("="*50)
print("Correlation Analysis")
print("="*50)

# Compute correlation matrix
corr_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# %%
# Pairplot to visualize relationships
print("\nPairplot of Numerical Variables:")
sns.pairplot(df[numerical_cols], diag_kind='kde')
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()

# %%
# Relationship between categorical and numerical variables
print("\nRelationship Between Categorical and Numerical Variables:")

for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=df, x=cat_col, y=num_col)
        plt.title(f'{num_col} by {cat_col}')
        plt.xticks(rotation=45)
        plt.show()

# %% [markdown]
# ## 4. Multivariate Analysis

# %%
# Stress level vs exercise time by BMI category
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Exercise_Time_Min', y='Stress_Level',
                hue='BMI_Category', style='Adequate_Sleep',
                palette='viridis', s=100)
plt.title('Stress Level vs Exercise Time by BMI Category and Sleep Adequacy')
plt.xlabel('Exercise Time (minutes)')
plt.ylabel('Stress Level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# %%
# Facet grid of sleep hours vs healthy meals by BMI category
g = sns.FacetGrid(df, col='BMI_Category', col_wrap=2, height=5)
g.map(sns.scatterplot, 'Healthy_Meals_Per_Day', 'Sleep_Hours_Per_Night',
      alpha=0.7)
g.set_axis_labels('Healthy Meals Per Day', 'Sleep Hours Per Night')
g.fig.suptitle('Sleep Hours vs Healthy Meals by BMI Category', y=1.02)
plt.show()

# %%
# 3D visualization (if you want to view in interactive mode)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

xs = df['Exercise_Time_Min']
ys = df['Sleep_Hours_Per_Night']
zs = df['BMI']

ax.scatter(xs, ys, zs, c=df['Stress_Level'], cmap='viridis', s=50)
ax.set_xlabel('Exercise Time (min)')
ax.set_ylabel('Sleep Hours')
ax.set_zlabel('BMI')
ax.set_title('3D View: Exercise, Sleep, BMI (colored by Stress Level)')
plt.show()

# %% [markdown]
# ## 5. Advanced Descriptive Statistics

# %%
# Group statistics by BMI category
print("="*50)
print("Group Statistics by BMI Category")
print("="*50)

group_stats = df.groupby('BMI_Category')[numerical_cols].agg(['mean', 'median', 'std', 'min', 'max'])
display(group_stats)

# %%
# Stress level distribution by sleep adequacy and BMI category
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='BMI_Category', y='Stress_Level', hue='Adequate_Sleep')
plt.title('Stress Level Distribution by BMI Category and Sleep Adequacy')
plt.xlabel('BMI Category')
plt.ylabel('Stress Level')
plt.legend(title='Adequate Sleep (≥7 hours)')
plt.show()

# %%
# Time series analysis (if date was available)
# This is just a placeholder to show what you might do with temporal data
print("\nNote: If the data had temporal components, we would include time series analysis here.")

# %% [markdown]
# ## 6. Key Insights Summary

# %%
# Calculate some key metrics
print("="*50)
print("Key Insights Summary")
print("="*50)

# Average metrics
avg_exercise = df['Exercise_Time_Min'].mean()
avg_sleep = df['Sleep_Hours_Per_Night'].mean()
avg_stress = df['Stress_Level'].mean()
avg_bmi = df['BMI'].mean()

print(f"Average Exercise Time: {avg_exercise:.1f} minutes per day")
print(f"Average Sleep Duration: {avg_sleep:.1f} hours per night")
print(f"Average Stress Level: {avg_stress:.1f}/10")
print(f"Average BMI: {avg_bmi:.1f}")

# Percentage with adequate sleep
pct_adequate_sleep = df['Adequate_Sleep'].mean() * 100
print(f"\nPercentage with adequate sleep (≥7 hours): {pct_adequate_sleep:.1f}%")

# BMI category distribution
bmi_dist = df['BMI_Category'].value_counts(normalize=True) * 100
print("\nBMI Category Distribution:")
display(bmi_dist)

# Most common stress level
common_stress = df['Stress_Level'].mode()[0]
print(f"\nMost common stress level: {common_stress}/10")

# Correlation between exercise and stress
corr_ex_stress = df['Exercise_Time_Min'].corr(df['Stress_Level'])
print(f"\nCorrelation between exercise time and stress level: {corr_ex_stress:.2f}")

# %%
# Save key statistics to a file
summary_stats = {
    'Average_Exercise_Min': avg_exercise,
    'Average_Sleep_Hours': avg_sleep,
    'Average_Stress_Level': avg_stress,
    'Average_BMI': avg_bmi,
    'Pct_Adequate_Sleep': pct_adequate_sleep,
    'Corr_Exercise_Stress': corr_ex_stress
}

summary_df = pd.DataFrame.from_dict(summary_stats, orient='index', columns=['Value'])
summary_df.to_csv('health_wellness_summary_stats.csv')

print("\nKey statistics saved to 'health_wellness_summary_stats.csv'")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from yellowbrick.cluster import KElbowVisualizer
import plotly.express as px

# Load the dataset
data = pd.read_csv('simulated_health_wellness_data.csv')

# Display basic info
print("Dataset Info:")
print(data.info())
print("\nFirst 5 rows:")
print(data.head())

# Exploratory Data Analysis
plt.figure(figsize=(12, 8))
sns.pairplot(data, diag_kind='kde')
plt.suptitle('Pairplot of Health Variables', y=1.02)
plt.show()

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()

# Data Preprocessing
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Determine optimal number of clusters using Elbow Method
plt.figure(figsize=(10, 6))
visualizer = KElbowVisualizer(KMeans(random_state=42), k=(2,10))
visualizer.fit(scaled_data)
visualizer.show()

# Silhouette analysis
silhouette_scores = []
for n_clusters in range(2, 10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(10, 6))
plt.plot(range(2, 10), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis')
plt.grid()
plt.show()

# Based on the plots, let's choose 4 clusters
n_clusters = 4

# K-Means Clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)
data['KMeans_Cluster'] = kmeans_labels

# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=n_clusters,
                                      metric='euclidean',
                                      linkage='ward')
hierarchical_labels = hierarchical.fit_predict(scaled_data)
data['Hierarchical_Cluster'] = hierarchical_labels

# Plot dendrogram
plt.figure(figsize=(15, 8))
dendrogram_plot = dendrogram(linkage(scaled_data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Dimensionality Reduction with PCA
pca = PCA()
pca.fit(scaled_data)

# Explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
         np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.grid()
plt.show()

# Use 2 principal components for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
data_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
data_pca['KMeans_Cluster'] = data['KMeans_Cluster']
data_pca['Hierarchical_Cluster'] = data['Hierarchical_Cluster']

# Plot clusters in PCA space
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='PC1', y='PC2', hue='KMeans_Cluster',
                data=data_pca, palette='viridis', s=100)
plt.title('K-Means Clusters in PCA Space')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')

plt.subplot(1, 2, 2)
sns.scatterplot(x='PC1', y='PC2', hue='Hierarchical_Cluster',
                data=data_pca, palette='viridis', s=100)
plt.title('Hierarchical Clusters in PCA Space')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')

plt.tight_layout()
plt.show()

# Interactive 3D plot (optional)
fig = px.scatter_3d(data, x='Exercise_Time_Min', y='Sleep_Hours_Per_Night', z='BMI',
                    color='KMeans_Cluster',
                    title='3D Cluster Visualization',
                    labels={'Exercise_Time_Min': 'Exercise (min)',
                            'Sleep_Hours_Per_Night': 'Sleep (hours)',
                            'BMI': 'BMI'})
fig.show()

# Cluster analysis
cluster_means = data.groupby('KMeans_Cluster').mean()
print("\nCluster Characteristics:")
print(cluster_means)

# Visualize cluster characteristics
plt.figure(figsize=(12, 6))
sns.heatmap(cluster_means.T, cmap='YlGnBu', annot=True, fmt='.1f')
plt.title('Cluster Characteristics (Mean Values)')
plt.xlabel('Cluster')
plt.ylabel('Variable')
plt.show()

# PCA component interpretation
pca_components = pd.DataFrame(pca.components_,
                             columns=data.columns[:-2],
                             index=['PC1', 'PC2'])
print("\nPCA Component Loadings:")
print(pca_components)

# Save results
data.to_csv('clustered_health_data.csv', index=False)
print("\nClustered data saved to 'clustered_health_data.csv'")
```
